```python
# Importing dependencies
import pandas as pd
```

```python
# Getting csv paths
data_01_2020 = "Data/JC-202001-citibike-tripdata.csv"
data_02_2020 = "Data/JC-202002-citibike-tripdata.csv"
data_03_2020 = "Data/JC-202003-citibike-tripdata.csv"
data_04_2020 = "Data/JC-202004-citibike-tripdata.csv"
data_05_2020 = "Data/JC-202005-citibike-tripdata.csv"
data_06_2020 = "Data/JC-202006-citibike-tripdata.csv"
data_07_2020 = "Data/JC-202007-citibike-tripdata.csv"
data_08_2020 = "Data/JC-202008-citibike-tripdata.csv"
data_09_2020 = "Data/JC-202009-citibike-tripdata.csv"
data_10_2020 = "Data/JC-202010-citibike-tripdata.csv"
data_11_2020 = "Data/JC-202011-citibike-tripdata.csv"
data_12_2020 = "Data/JC-202012-citibike-tripdata.csv"
```

```python
# Creating a file list
file_list = [data_01_2020, data_02_2020, data_03_2020, data_04_2020, data_05_2020, data_06_2020, 
             data_07_2020, data_08_2020, data_09_2020, data_10_2020, data_11_2020, data_12_2020]

# Creating empty list to append dfs
data_list = []

# Looping through files to append data into new list
for file in file_list:
    data_list.append(pd.read_csv(file))

# Combining dfs into a single one
data_2020 = pd.concat(data_list, sort = False)

# Previewing final df
data_2020.head()
```

```python
# Looking for possible outliers
data_2020.describe()
```

```python
# Discarding riders born before 1920
final_df = data_2020.loc[data_2020["birth year"] > 1920, :]

final_df.describe()
```

```python
# Save final df to csv file
final_df.to_csv("data/bikedata_2020.csv", index=False)
```
