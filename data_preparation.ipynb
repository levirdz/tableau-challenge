```python
# Importing dependencies
import pandas as pd
```

```python
# Getting csv paths
data_01_2019 = "Data/JC-201901-citibike-tripdata.csv"
data_02_2019 = "Data/JC-201902-citibike-tripdata.csv"
data_03_2019 = "Data/JC-201903-citibike-tripdata.csv"
data_04_2019 = "Data/JC-201904-citibike-tripdata.csv"
data_05_2019 = "Data/JC-201905-citibike-tripdata.csv"
data_06_2019 = "Data/JC-201906-citibike-tripdata.csv"
data_07_2019 = "Data/JC-201907-citibike-tripdata.csv"
data_08_2019 = "Data/JC-201908-citibike-tripdata.csv"
data_09_2019 = "Data/JC-201909-citibike-tripdata.csv"
data_10_2019 = "Data/JC-201910-citibike-tripdata.csv"
data_11_2019 = "Data/JC-201911-citibike-tripdata.csv"
data_12_2019 = "Data/JC-201912-citibike-tripdata.csv"
```

```python
# Creating a file list
file_list = [data_01_2019, data_02_2019, data_03_2019, data_04_2019, data_05_2019, data_06_2019, data_07_2019, 
             data_08_2019, data_09_2019, data_10_2019, data_11_2019, data_12_2019]

# Creating empty list to append dfs
data_list = []

# Looping through files to append data into new list
for file in file_list:
    data_list.append(pd.read_csv(file))

# Combining dfs into a single one
data_2019 = pd.concat(data_list, sort = False)

# Previewing final df
data_2019.head()
```
